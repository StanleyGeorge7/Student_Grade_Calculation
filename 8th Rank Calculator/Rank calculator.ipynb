```python
no_of_students=int(input('Enter the no of students:'))
subjects=['Tamil','English','maths','science','social']
entire_class=[]
for i in  range(1,(no_of_students+1)):
    student_info=[]
    roll_number=int(input('Enter the Roll number'))
    student_info.append(roll_number)
    total_marks=0
    for subject in subjects: 
        marks=int(input(f'Enter the FA for {subject}'))
        student_info.append(marks)
        total_marks+=marks
    student_info.append(total_marks)
    entire_class.append(student_info)
    
student_info
```

```python
import pandas as pd
pd.options.display.max_columns = 50
df=pd.DataFrame(entire_class,columns=['Roll Number','Tamil','English','Maths','science','social','Total'])
df['Roll Number'].replace(20,1,inplace=True)
df
```

```python
df["Rank"] = df["Total"].rank(ascending=0)
df['Rank']= df["Rank"].astype('int')
```
