```python
no_of_students=int(input('Enter the no of students:'))
subjects=['Tamil','English','maths','science','social']
entire_class=[]
for i in  range(1,(no_of_students+1)):
    student_info=[]
    roll_number=int(input('Enter the Roll number'))
    student_info.append(roll_number)
    for subject in subjects: 
        fa=int(input(f'Enter the FA for {subject}'))
        student_info.append(fa)
        if(fa>=37) and (fa<=40):
            student_info.append('A1')
        elif(fa>=33) and (fa<=36):
            student_info.append('A2')
        elif(fa>=29) and (fa<=32):
            student_info.append('B1')
        elif(fa>=25) and (fa<=28):
            student_info.append('B2')
        elif(fa>=21) and (fa<=24):
            student_info.append('C1')
        elif(fa>=17) and (fa<=20):
            student_info.append('C2')
        elif(fa>=13) and (fa<=16):
            student_info.append('D')
        elif(fa>=9) and (fa<=12):
            student_info.append('E1')
        elif (fa<=8):
            student_info.append('E2')
        sa=int(input(f'Enter the SA for {subject}'))
        student_info.append(sa)
        if(sa>=55) and (sa<=60):
            student_info.append('A1')
        elif(sa>=49) and (sa<=54):
            student_info.append('A2')
        elif(sa>=43) and (sa<=48):
            student_info.append('B1')
        elif(sa>=37) and (sa<=42):
            student_info.append('B2')
        elif(sa>=31) and (sa<=36):
            student_info.append('C1')
        elif(sa>=25) and (sa<=30):
            student_info.append('C2')
        elif(sa>=19) and (sa<=24):
            student_info.append('D')
        elif(sa>=13) and (sa<=18):
            student_info.append('E1')
        elif (sa<=12):
            student_info.append('E2')
        total=fa+sa
        student_info.append(total)
        if(total>=91) and (total<=100):
            student_info.append('A1')
        elif(total>=81) and (total<=90):
            student_info.append('A2')
        elif(total>=71) and (total<=80):
            student_info.append('B1')
        elif(total>=61) and (total<=70):
            student_info.append('B2')
        elif(total>=51) and (total<=60):
            student_info.append('C1')
        elif(total>=41) and (total<=50):
            student_info.append('C2')
        elif(total>=33) and (total<=40):
            student_info.append('D')
        elif(total>=21) and (total<=32):
            student_info.append('E1')
        elif (total<=20):
            student_info.append('E2')
    entire_class.append(student_info)
        
        
          
        
```

```python
import pandas as pd
pd.options.display.max_columns = 50
df=pd.DataFrame(entire_class,columns=['Roll Number','THAMIZH FA','THAMIZH FA GRADE','THAMIZH SA','THAMIZH SA GRADE','THAMIZH TOTAL','THAMIZH TOTAL GRADE',
                                       'ENGLISH FA','ENGLISH FA GRADE','ENGLISH SA','ENGLISH SA GRADE','ENGLISH TOTAL','ENGLISH TOTAL GRADE',
                                       'MATHS FA','MATHS FA GRADE','MATHS SA','MATHS SA GRADE','MATHS TOTAL','MATHS TOTAL GRADE',
                                       'SCIENCE FA','SCIENCE FA GRADE','SCIENCE SA','SCIENCE SA GRADE','SCIENCE TOTAL','SCIENCE TOTAL GRADE',
                                       'SOCIAL FA','SOCIAL FA GRADE','SOCIAL SA','SOCIAL SA GRADE','SOCIAL TOTAL','SOCIAL TOTAL GRADE'])
df
```
